*args dan **kwargs

```python
def jumlahkan(*args):
    return sum(args)
def biodata(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")
print("Total:", jumlahkan(1, 2, 3, 4, 5))
biodata(Nama="Budi", Umur=20, Prodi="TRK")
```

Decorator

```python
def decorator_sapa(fungsi):
    def wrapper():
        print("Sebelum fungsi dijalankan...")
        fungsi()
        print("Setelah fungsi dijalankan...")
    return wrapper
@decorator_sapa
def sapa():
    print("Halo, selamat datang!")
sapa()
```

Penugasan 9: studi kasus nyata - program nilai

```python

def input_nilai():
    n = int(input("Masukkan jumlah mahasiswa: "))
    nilai = []
    for i in range(n):
        data = float(input(f"Masukkan nilai mahasiswa ke-{i+1}: "))
        nilai.append(data)
    return nilai

def hitung_rata(nilai):
    return sum(nilai) / len(nilai)

def status_kelulusan(rata):
    if rata >= 60:
        return "Lulus"
    else:
        return "Tidak Lulus"

nilai_akhir = lambda tugas, ujian: (0.4 * tugas) + (0.6 * ujian)

print("=== Program Nilai Mahasiswa ===")
daftar_nilai = input_nilai()
rata = hitung_rata(daftar_nilai)
print(f"\nRata-rata nilai: {rata:.2f}")
print("Status kelulusan:", status_kelulusan(rata))

tugas = float(input("\nMasukkan nilai tugas: "))
ujian = float(input("Masukkan nilai ujian: "))
akhir = nilai_akhir(tugas, ujian)
print(f"Nilai akhir berdasarkan bobot = {akhir:.2f}")
```
