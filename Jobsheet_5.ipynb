Indexing dan Slicing


```python
teks = "Politeknik Negeri Semarang"

# Indexing
print(teks[0])     # Huruf pertama
print(teks[-1])    # Huruf terakhir

# Slicing
print(teks[0:10])  # Karakter dari index 0 sampai 9
print(teks[11:17]) # Karakter "Negeri"
print(teks[:9])    # Dari awal sampai index 8
print(teks[10:])   # Dari index 10 sampai akhir
```

Method bawaan string

```python
teks = " Politeknik Negeri Semarang "

print(teks.strip())      # Hilangkan spasi di awal & akhir
print(teks.upper())      # Jadi huruf besar semua
print(teks.lower())      # Jadi huruf kecil semua
print(teks.replace("Semarang", "Bandung"))  # Ganti kata
print(teks.count("e"))   # Hitung jumlah huruf 'e'
```

Casting data

```python
angka_str = input("Masukkan angka (string): ")

# Ubah ke integer dan hitung kuadratnya
angka_int = int(angka_str)
print(f"Kuadrat dari {angka_int} adalah {angka_int ** 2}")

# Ubah ke float dan tampilkan dengan 2 angka di belakang koma
angka_float = float(angka_str)
print(f"Sebagai float: {angka_float:.2f}")
```

Studi kasus : analisis teks

```python
kalimat = input("Masukkan kalimat: ")

kata_list = kalimat.split()

# Hitung jumlah kata
jumlah_kata = len(kata_list)

# Cari kata terpanjang
kata_terpanjang = max(kata_list, key=len)

# Urutkan kata secara alfabet
kata_urut = sorted(kata_list)

print(f"Jumlah kata       : {jumlah_kata}")
print(f"Kata terpanjang   : {kata_terpanjang}")
print(f"Daftar kata urut  : {kata_urut}")
```
