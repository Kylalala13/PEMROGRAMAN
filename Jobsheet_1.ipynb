menghitung luas segitiga

```python
import math
a = float(input("Masukkan sisi a: "))
b = float(input("Masukkan sisi b: "))
c = float(input("Masukkan sisi c: "))
s = (a + b + c) / 2
luas = math.sqrt (s * (s-a)*(s-b)*(s-c))
print ("luas segitiga =", luas)
```
