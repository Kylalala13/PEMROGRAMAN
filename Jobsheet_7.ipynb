Studi kasus dictionary : daftar nilai

```python
nilai_mahasiswa = {
    "Andi": 85,
    "Budi": 92,
    "Cici": 90
}
for nama, nilai in nilai_mahasiswa.items():
    print(f"{nama}: {nilai}")
rata = sum(nilai_mahasiswa.values()) / len(nilai_mahasiswa)
print(f"Rata-rata nilai: {rata:2f}")
```

Stusi kasus set - klub mahasiswa

```python
# Studi Kasus Set – Klub Mahasiswa

# Membuat set anggota klub
klub_A = {"Andi", "Budi", "Cici"}
klub_B = {"Cici", "Deni", "Eka"}

# Operasi himpunan
semua_anggota = klub_A.union(klub_B)          # Gabungan
anggota_kedua_klub = klub_A.intersection(klub_B)  # Irisan
hanya_klub_A = klub_A.difference(klub_B)          # Selisih (A - B)

# Menampilkan hasil
print("Daftar Seluruh Anggota (Union):", semua_anggota)
print("Anggota di Kedua Klub (Intersection):", anggota_kedua_klub)
print("Anggota Hanya di Klub A (Difference):", hanya_klub_A)
```
