Pengenalan variabel & tipe data


```python
#Integer
umur = 20
print ("Umur:", umur, "→", type (umur))

#Float
tinggi = 170,5
print ("tinggi:", tinggi, "→", type (tinggi))

#String
nama = "andi"
print  ("nama:", nama, "→", type (nama))

#Boolean
is_mahasiswa = True
print ("Status Mahasiswa:", is_mahasiswa, "→", type (is_mahasiswa))

#List
nilai = [80, 85, 90]
print("Nilai:", nilai, "→", type(nilai))

#Tuple
koordinat = (10, 20)
print("Koordinat:", koordinat, "→", type(koordinat))

#Dictionary
mahasiswa = {"nama": "Andi", "umur": 20, "prodi": "TRK"}
print("Data Mahasiswa:", mahasiswa, "→", type(mahasiswa))

# Set
hobi = {"membaca", "olahraga", "musik"}
print("Hobi:", hobi, "→", type(hobi))
```

Program Kasir Sederhana

```python
barang = input("Masukkan nama barang: ")
harga = float(input("Masukkan harga barang:"))
jumlah = int(input("Masukkan jumlah barang"))

total = harga * jumlah

# Aturan diskon
if total >= 500000:
    diskon = 0.2
elif total >= 250000:
    diskon = 0.1
else:
    diskon = 0

potongan = total * diskon
bayar = total - potongan

print("\n===== Struktur Belanja =====")
print(f"Barang : {barang}")
print(f"Jumlah : {jumlah}")
print(f"Total : Rp {total}")
print(f"Diskon : {diskon * 100}%")
print(f"Bayar : Rp {bayar}")
```

Operasi Relasional

```python
# Program: Operator Relasional

# Membaca dua input bilangan dari pengguna
a = float(input("Masukkan bilangan pertama: "))
b = float(input("Masukkan bilangan kedua: "))

# Menampilkan hasil perbandingan
print("\n=== Hasil Perbandingan ===")
print(f"{a} == {b}  →  {a == b}")   # Sama dengan
print(f"{a} != {b}  →  {a != b}")   # Tidak sama dengan
print(f"{a} <  {b}  →  {a < b}")    # Kurang dari
print(f"{a} >  {b}  →  {a > b}")    # Lebih dari
print(f"{a} <= {b}  →  {a <= b}")   # Kurang dari atau sama dengan
print(f"{a} >= {b}  →  {a >= b}")   # Lebih dari atau sama dengan
```

Match-Case

```python
menu = int(input("Pilih menu (1-3): "))

match menu:
  case 1:
    print("Anda memilih Nasi Goreng")
  case 2:
    print("Anda memilih Mie Ayam")
  case 3:
    print("Anda memilih Soto Ayam")
  case _:
    print("Menu tidak tersedia")
```

Menghitung Tarif Parkir

```python
# Meminta input durasi parkir dari pengguna dalam jam
durasi_parkir = int(input("Masukan durasi parkir (dalam jam): "))

#Menghitung tarif parkir
if durasi_parkir <= 2:
    tarif = 5000

else:
    # Tarif untuk jam-jam berikutnya
    tarif = 5000 + (durasi_parkir - 2) * 2000

# Menerapkan tarif maksimal
if tarif > 20000:
    tarif = 20000

print(f"Tarif parkir Anda adalah: Rp {tarif}")
```

Studi Kasus Nyata – Penentuan Diskon

```python
# Meminta input nama barang, harga, dan jumlah
barang = input("Masukan nama barang: ")
harga = float(input("Masukan harga barang: "))
jumlah = int(input("Masukan jumlah barang: "))

#Menghitung total belanja
total = harga * jumlah

# Menentukan diskon berdasarkan total belanja
if total >= 500000:
    diskon = 0.2
elif total >= 250000:
    diskon = 0.1
else:
    diskon = 0

# Menghitung jumlah bayar setelah diskon
bayar = total - (total * diskon)

print("\n===== Struk Belanja =====")
print(f"Barang : {barang}")
print(f"Jumlah : {jumlah}")
print(f"Total : Rp {total}")
print(f"Diskon : {diskon * 100}%")
print(f"Bayar : Rp {bayar}")
```
