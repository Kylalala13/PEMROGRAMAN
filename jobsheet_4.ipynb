For dengan range

```python
for i in range(1, 6):
    print(i)
```

While ( Hitung mundur )

```python
n = 5
while n > 0:
    print(n)
    n -= 1
print("Selesai!")
```

Nested Loop ( Segitiga bintang )

```python
for i in range(1, 6):
    for j in range(1, i + 1):
        print("*", end="")
    print()  # ganti baris
```

Break - Continue

```python
# Break
for i in range(1, 11):
    if i == 6:
        break
    print("Break :", i)
```

```python
# Continue
for i in range(1, 6):
    if i == 3:
        continue
    print("Continue :", i)
```

For pada Dictionary

```python
data = {"Nama": "Andi", "Umur": 20, "Prodi": "TRK"}

for key, value in data.items():
    print(key, ":", value)
```

For dengan Enumerate

```python
nama = ["Andi", "Budi", "Cici"]

for index, value in enumerate(nama, start=1):
    print(index, "-", value)
```

Kasir multi item

```python
# Data barang
barang = [
    {"nama": "Buku", "harga": 50000, "jumlah": 2},
    {"nama": "Pulpen", "harga": 5000, "jumlah": 3}
]

total = 0
print("===== Struk Pembelian =====")
for item in barang:
    subtotal = item["harga"] * item["jumlah"]
    total += subtotal
    print(f"{item['nama']} (x{item['jumlah']}) = Rp{subtotal:,}")

print("============================")
print(f"Total Bayar: Rp{total:,}")
```

Hitung faktorial

```python
# Program menghitung faktorial dengan while
n = int(input("Masukkan bilangan: "))

faktorial = 1
i = 1

while i <= n:
    faktorial *= i
    i += 1

print(f"Faktorial dari {n} adalah {faktorial}")
```

Pola segitiga terbalik

```python
# Program Pola Segitiga Terbalik

tinggi = 5

for i in range(tinggi):         # loop baris
    for j in range(tinggi - i): # loop kolom
        print("*", end="")
    print()  # pindah ke baris baru
```

Mencari bilangan prima dengan break

```python
# Program Cek Bilangan Prima dengan Break

bilangan = int(input("Masukkan bilangan: "))

if bilangan > 1:
    for i in range(2, bilangan):
        if bilangan % i == 0:
            print(bilangan, "bukan bilangan prima")
            break
    else:
        print(bilangan, "adalah bilangan prima")
else:
    print(bilangan, "bukan bilangan prima")
```

Tampilkan hanya bilangan genap dengn continue

```python
# Program Menampilkan Bilangan Genap dengan Continue

for i in range(1, 21):
    if i % 2 != 0:
        continue  # lewati bilangan ganjil
    print(i)
```

Studi kasus - kasis diskon

```python
# Program Kasir Diskon Sederhana

print("=== Program Kasir Diskon ===")
total = 0
barang_list = []

while True:
    nama = input("Masukkan nama barang (ketik 'selesai' untuk berhenti): ")
    if nama.lower() == "selesai":
        break

    harga = float(input(f"Masukkan harga {nama}: Rp "))
    jumlah = int(input(f"Masukkan jumlah {nama}: "))

    subtotal = harga * jumlah
    total += subtotal
    barang_list.append((nama, harga, jumlah, subtotal))
    print(f"Subtotal {nama}: Rp {subtotal:,.0f}\n")

# Hitung diskon
if total >= 200000:
    diskon = total * 0.10
else:
    diskon = 0

total_bayar = total - diskon

# Tampilkan struk belanja
print("\n===== STRUK BELANJA =====")
for i, (nama, harga, jumlah, subtotal) in enumerate(barang_list, start=1):
    print(f"{i}. {nama} ({jumlah} x Rp{harga:,.0f}) = Rp{subtotal:,.0f}")

print("--------------------------")
print(f"Total Belanja : Rp{total:,.0f}")
print(f"Diskon (10%)  : Rp{diskon:,.0f}")
print(f"Total Bayar   : Rp{total_bayar:,.0f}")
print("==========================")
print("Terima kasih telah berbelanja!")
```
