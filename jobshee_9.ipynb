Penugasan 4: Modul buatan sendiri

```python
%%writefile konversisuhu.py

def celcius_ke_fahrenheit(c):
  fahrenheit = (c * 9/5) + 32
  return fahrenheit

def fahrenheit_ke_celcius(f):
  celcius = (f - 32) * 5/9
  return celcius
```

```python
import konversisuhu

suhu_c = int(input("Masukan suhu dalam celcius = "))
suhu_f = konversisuhu.celcius_ke_fahrenheit(suhu_c)
print(f"{suhu_c}°C = {suhu_f}°F")

suhu_f2 = int(input("Masukan suhu dalam fahrenheit = "))
suhu_c2 = konversisuhu.fahrenheit_ke_celcius(suhu_f2)
print(f"{suhu_f2}°F = {suhu_c2}°C")
```

Penugasan 6: studi kasus mini

```python
%%writefile laporan.py

def tampilkan_skor(benar, salah):
  """Menampilkan ringkasan skor"""
  total = benar + salah
  skor = (benar / total) * 100 if total > 0 else 0 # Handle division by zero

  print("HASIL KUIS")
  print(f"Benar: {benar}")
  print(f"Salah: {salah}")
  print(f"Skor: {skor:.0f}")
```

```python
import random
import laporan

jumlah_soal = 5

benar = 0
salah = 0

print("KUIS PENJUMLAHAN")

for i in range(1, jumlah_soal + 1):
  angka1 = random.randint(1, 20)
  angka2 = random.randint(1, 20)
  jawaban_benar = angka1 + angka2

  print(f"\nSoal {i}: {angka1} + {angka2} = ?")
  jawaban = int(input("Jawaban: "))

  if jawaban == jawaban_benar:
    print("Benar!")
    benar += 1
  else:
    print(f"Salah! Jawaban yang benar adalah {jawaban_benar}")
    salah += 1

laporan.tampilkan_skor(benar, salah)
```
