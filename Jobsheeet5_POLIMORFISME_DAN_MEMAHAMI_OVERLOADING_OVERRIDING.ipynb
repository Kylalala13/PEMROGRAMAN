Praktikum 1: Polimorfisme dengan Inheritance dan Overriding

```python
import inspect

# Kelas Induk
class Burung:
    def __init__(self, nama): # Fixed: Changed _init_ to __init__
        self.nama = nama

    def terbang(self):
        print(f"{self.nama} terbang dengan cara umum.")

    def bersuara(self):
        print(f"{self.nama} mengeluarkan suara burung.")


# Kelas Anak 1
class Elang(Burung):
    def __init__(self, nama, rentang_sayap):
        super().__init__(nama)
        self.rentang_sayap = rentang_sayap

    # Override
    def bersuara(self): # Fixed: Removed the first duplicate bersuara method
        print(f"{self.nama} berteriak nyaring!")


# Kelas Anak 2
class Pipit(Burung):
    def __init__(self, nama, warna_bulu):
        super().__init__(nama)
        self.warna_bulu = warna_bulu

    # Override
    def terbang(self):
        print(f"{self.nama} terbang cepat di antara pepohonan.")

    # override
    def bersuara(self):
        print(f"{self.nama} berkicau merdu: Cit Cit!")

# Fungsi yang memanfaatkan polimorfisme
def demonstrasi_aksi_burung(daftar_burung):
    print("\nAksi Burung:")
    for burung in daftar_burung:
        print(f"-- Aksi untuk {burung.nama} --")
        # Panggilan metode yang sama, tapi perilaku berbeda tergantung objek
        burung.terbang()
        burung.bersuara()
        print("-" * 15)

# --- Kode utama ---
if __name__ == "__main__":
    elang_jawa = Elang("Elang Jawa", 1.5)
    pipit_gereja = Pipit("Pipit Gereja", "Coklat")
    burung_aneh = Burung("Burung Misterius") # Objek dari kelas induk

    koleksi_burung = [elang_jawa, pipit_gereja, burung_aneh]
    demonstrasi_aksi_burung(koleksi_burung)
```

Praktikum 2: Polimorfisme dengan Duck Typing

```python
class Kucing:
    def bersura(self):
        print("Kucing: Meow!")

class Anjing:
    def bersuara(self):
        print("Anjing: Guk guk!")

class Bebek:
    def bersuara(self):
        print("Bebek: Kwek kwek!")

class Mobil: # kelas ini TIDAK punya metode suara()
    def jalankan(self):
        print("Mobil: Brummm!")

# Fungsi ini tidak peduli tipe objeknya,
# asal punya metode .bersuara()
def buat_suara(objek_yang_bisa_bersuara):
    try:
        # Memanggil metode 'bersuara' jika ada
        objek_yang_bisa_bersuara.bersuara()
    except AttributeError:
        # Menangani jika objek tidak punya metode 'bersuara'
        print(f"Objek {type(objek_yang_bisa_bersuara).__name__} tidak bisa bersuara.")

# --- Kode utama ---
if __name__ == "__main__":
    kucing = Kucing()
    anjing = Anjing()
    bebek = Bebek()
    mobil = Mobil()

    daftar_objek = [kucing, anjing, bebek, mobil] # Fixed: Changed kucing1 to kucing, anjing1 to anjing, etc.
    print("Demonstrasi Duck Typing:")
    for item in daftar_objek:
        buat_suara(item)
```

Praktikum	3:	Simulasi Overloading	dalam Metode Kelas dengan *args

```python
class KalkulatorSederhana:
    def __init__(self, nama="Kalkulator"):
        self.nama = nama
        print(f"{self.nama} siap digunakan.")

    def tambah(self, *args):
        print(f"\nMemanggil metode tambah dengan argumen: {args}")
        if not args:
            print("Tidak ada angka untuk dijumlahkan.")
            return 0

        total = 0
        valid_input = True
        for angka in args:
            if isinstance(angka, (int, float)):
                total += angka
            else:
                print(f"Peringatan: Argumen '{angka}' bukan angka dan akan diabaikan.")
                valid_input = False

        if valid_input:
            print(f"Hasil penjumlahan: {total}")
        else:
            print(f"Hasil penjumlahan (dengan beberapa input diabaikan): {total}")
        return total

# --- Kode Utama ---
if __name__ == "__main__":
    calc = KalkulatorSederhana("Calc-01")
    print("\n--- Percobaan Penjumlahan ---")
    calc.tambah(5, 10)
    calc.tambah(2, 3, 5, 10)
    calc.tambah(100)
    calc.tambah() # Tanpa argumen
    calc.tambah(1, 2, "tiga", 4, 5.5)
```

Praktikum	4:	Simulasi Overloading	dalam Metode Kelas dengan *args dan **kwargs

```python
class DataProcessor:
    def __init__(self, processor_id):
        self.processor_id = processor_id
        print(f"Data Processor {self.processor_id} siap.")

    def process(self, *args, **kwargs):
        print(f"\n--- {self.processor_id} Memproses Data ---")

        # Memproses argumen posisi
        if args:
            print("Argumen Posisi Diterima:")
            for i, arg in enumerate(args):
                print(f" args[{i}]: {arg} (tipe: {type(arg).__name__})")
        else:
            print("Tidak ada argumen posisi.")

        # Memproses argumen kata kunci
        if kwargs:
            print("Argumen Kata Kunci Diterima:")
            for kunci, nilai in kwargs.items():
                print(f" {kunci}: {nilai} (tipe: {type(nilai).__name__})")
        else:
            print("Tidak ada argumen kata kunci.")

# --- Kode Utama ---
if __name__ == "__main__":
    processor1 = DataProcessor("DP-001")

    print("\nPanggilan 1: Tanpa argumen tambahan")
    processor1.process()

    print("\nPanggilan 2: Hanya argumen posisi")
    processor1.process(100, "Status OK", 99.9, False)

    print("\nPanggilan 3: Hanya argumen kata kunci")
    processor1.process(user="admin", level=5, mode="verbose")

    print("\nPanggilan 4: Kombinasi argumen posisi dan kata kunci")
    processor1.process("Task-A", "Task-B", status="Running", priority="High", thread_id=54321)
```

Praktikum	5:	Kombinasi	Polimorfisme (Inheritance & Duck Typing)

```python
import math

class Bentuk:
    def nama_bentuk(self):
        return "Bentuk Generik"

    def hitung_luas(self):
        raise NotImplementedError("Subclass harus mengimplementasikan metode ini")

class Lingkaran(Bentuk):
    def __init__(self, radius):
        self.radius = radius
    def nama_bentuk(self):
        return "Lingkaran"
    def hitung_luas(self):
        return math.pi * (self.radius ** 2)

class Persegi(Bentuk):
    def __init__(self, sisi):
        self.sisi = sisi
    def nama_bentuk(self):
        return "Persegi"
    def hitung_luas(self):
        return self.sisi * self.sisi

class TaplakMeja: # Tidak terkait inheritance dengan Bentuk
    def __init__(self, panjang, lebar):
        self.panjang = panjang
        self.lebar = lebar
    def hitung_luas(self):
        return self.panjang * self.lebar

def tampilkan_info_luas(objek_dengan_luas):
    print("-" * 20)
    try:
        luas = objek_dengan_luas.hitung_luas()
        try:
            nama = objek_dengan_luas.nama_bentuk()
        except AttributeError:
            nama = type(objek_dengan_luas).__name__

        print(f"Objek: {nama}")
        print(f"Luas : {luas:.2f}")
    except AttributeError:
        print(f"Objek {type(objek_dengan_luas).__name__} tidak dapat dihitung luasnya.")
    except NotImplementedError:
        print(f"Metode 'hitung_luas' belum diimplementasi untuk {type(objek_dengan_luas).__name__}.")

# --- Kode Utama ---
if __name__ == "__main__":
    daftar_item = [Lingkaran(7), Persegi(5), TaplakMeja(1.5, 0.8), Bentuk(), "Ini string"]
    print("Menampilkan Info Luas (Polimorfisme Campuran):")
    for item in daftar_item:
        tampilkan_info_luas(item)
```

Praktikum	6:	Kontrol	Perilaku	Polimorfik dengan isinstance()

```python
class Burung:
    def __init__(self, nama):
        self.nama = nama
    def bersuara(self):
        print(f"{self.nama} bersuara.")

class Elang(Burung):
    def bersuara(self):
        print(f"{self.nama}: Kaaaak!")
    def berburu(self):
        print(f"{self.nama} sedang mengincar mangsa dari langit.")

class Pipit(Burung):
    def bersuara(self):
        print(f"{self.nama}: Cit cit!")
    def membangun_sarang(self):
        print(f"{self.nama} sedang mengumpulkan jerami untuk sarang.")

def interaksi_dengan_burung(burung):
    print(f"\nInteraksi dengan {burung.nama}:")
    burung.bersuara() # Polimorfisme murni

    # Perilaku khusus berdasarkan tipe
    if isinstance(burung, Elang):
        burung.berburu()
    elif isinstance(burung, Pipit):
        burung.membangun_sarang()

# --- Kode Utama ---
if __name__ == "__main__":
    elang = Elang("Elang Emas")
    pipit = Pipit("Pipit Kecil")

    interaksi_dengan_burung(elang)
    interaksi_dengan_burung(pipit)
```

```python
import abc

# 1. & 2. Membuat Kelas Induk 'Komputer' dengan Metode Abstrak/Umum
class Komputer:
    def __init__(self, merk, ram, processor):
        self.merk = merk
        self.ram = ram
        self.processor = processor

    def tampilkan_spesifikasi(self):
        # Akan di-override oleh kelas anak
        pass

    # 5. Simulasi Overloading menggunakan *args
    def install_software(self, *daftar_software):
        print(f"\n[Sistem {self.merk}] Memulai proses instalasi...")
        if not daftar_software:
            print("Tidak ada software yang dipilih untuk diinstal.")
        else:
            for sw in daftar_software:
                print(f"--> Berhasil menginstal: {sw}")

# 3. Membuat Kelas Anak 'Laptop' dan 'Desktop' (Method Overriding)
class Laptop(Komputer):
    def __init__(self, merk, ram, processor, berat_kg):
        super().__init__(merk, ram, processor)
        self.berat_kg = berat_kg

    def tampilkan_spesifikasi(self):
        print(f"Spesifikasi Laptop {self.merk}:")
        print(f"- Processor: {self.processor}")
        print(f"- RAM      : {self.ram} GB")
        print(f"- Berat    : {self.berat_kg} kg")

class Desktop(Komputer):
    def __init__(self, merk, ram, processor, tipe_casing):
        super().__init__(merk, ram, processor)
        self.tipe_casing = tipe_casing

    def tampilkan_spesifikasi(self):
        print(f"Spesifikasi Desktop {self.merk}:")
        print(f"- Processor: {self.processor}")
        print(f"- RAM      : {self.ram} GB")
        print(f"- Casing   : {self.tipe_casing}")

# 4. Fungsi Polimorfik (Duck Typing / Polymorphism)
def cetak_semua_spesifikasi(list_komputer):
    print("=== DAFTAR SPESIFIKASI PERANGKAT ===")
    for perangkat in list_komputer:
        perangkat.tampilkan_spesifikasi()
        print("-" * 35)

# 6. Kode Utama (Main Program)
if __name__ == "__main__":
    # Inisialisasi objek Laptop dan Desktop
    my_laptop = Laptop("ASUS ROG", 16, "AMD Ryzen 9", 2.1)
    my_desktop = Desktop("Dell Precision", 32, "Intel i9", "Tower Case")

    # Masukkan ke dalam list
    daftar_perangkat = [my_laptop, my_desktop]

    # Panggil fungsi polimorfik untuk mencetak spesifikasi
    cetak_semua_spesifikasi(daftar_perangkat)

    # Demonstrasi simulasi overloading (Langkah 5)
    # Contoh 1: Menginstal satu software
    my_laptop.install_software("VS Code")

    # Contoh 2: Menginstal banyak software sekaligus
    my_desktop.install_software("Python 3.12", "PyCharm", "Docker", "PostgreSQL")

    # Contoh 3: Tanpa argumen
    my_laptop.install_software()
```
