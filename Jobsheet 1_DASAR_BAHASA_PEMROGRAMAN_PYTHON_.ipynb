PRAKTIKUM 1: TIPE DATA DASAR

```python
# 1. Number(int, float, complex)
# integer (int)
angka_int = 10
print("angka_int =", angka_int, "->", type(angka_int))

# Float
angka_float = 3.14
print("angka_float =", angka_float, "->", type(angka_float))

# Complex
angka_complex = 2 + 3j
print("angka_complex =", angka_complex, "->", type(angka_complex))

# 2. Boolean
is_active = True
print("is_active =", is_active, "->", type(is_active))

# 3. String
teks = "True"
print("teks =", teks, "->", type(teks))

# 4. List
# List adalah tipe data terurut dan tapat diubah (mutable)
daftar_angka = [1, 2, 3, 4, 5]
print("daftar_angka =", daftar_angka, "->", type(daftar_angka))

# 5. Tuple
# Tuple adalah tipe data terurut tetapi tidak dapat diubah (immutable)
koordinat = (10, 20)
print("koordinat =", koordinat, "->", type(koordinat))

# 6. Dictionary
# Dictionary menyimpan data dalam pasangan key-value
data_mahasiswa = {
    "nama" : "Andi",
    "nim" : "A11.2022.12345",
    "jurusan" : "Teknik Informatika"
}
print("data_mahasiswa =", data_mahasiswa, "->", type(data_mahasiswa))

# 7. Set (Unik & Tidak Terurut)
himpunan_angka = {1, 2, 3, 2, 1}
print("himpunan_angka =", himpunan_angka, "->", type(himpunan_angka))

# 8. Konversi Tipe Data
nilai_str = "100"
print("\nnilai_str =", nilai_str, "->", type(nilai_str))
nilai_int = int(nilai_str) # konversi string ke integer
print("nilai_int =", nilai_int, "->", type(nilai_int))
```

PRAKTIKUM 2: VARIABEL DAN OPERASI DASAR

```python
# 1. Pendeklarasian Variabel
nama = "Budi"
umur = 20
tinggi = 170.5
is_student = True

print("Nama =", nama)
print("Umur =", umur)
print("Tinggi =", tinggi, "cm")
print("Mahasiswa =", is_student)

# 2. Operasi Aritmetika
a = 10
b = 3

penjumlahan = a + b
pengurangan = a - b
perkalian = a * b
pembagian = a / b          # Hasil float
pembagian_bulat = a // b   # Hasil bulat
modulus = a % b            # Sisa bagi
pangkat = a ** b           # Pemangkatan

print("\nOPERASI ARITMETIKA")
print("a =", a, ", b =", b)
print("Penjumlahan =", penjumlahan)
print("Pengurangan =", pengurangan)
print("Perkalian =", perkalian)
print("Pembagian =", pembagian)
print("Pembagian Bulat =", pembagian_bulat)
print("Modulus =", modulus)
print("Pangkat =", pangkat)

# 3. Operasi Perbandingan
print("\nOPERASI PERBANDINGAN")
print("a > b =", a > b)
print("a < b =", a < b)
print("a == b =", a == b)
print("a != b =", a != b)
print("a >= b =", a >= b)
print("a <= b =", a <= b)

# 4. Operasi Logika
x = True
y = False

print("\nOPERASI LOGIKA")
print("x =", x, ", y =", y)
print("x and y =", x and y)
print("x or y =", x or y)
print("not x =", not x)

# 5. Penggunaan dalam Percabangan Sederhana
if a > b and b > 0:
    print("\nKondisi terpenuhi: a lebih besar dari b, dan b masih positif.")
else:
    print("\nKondisi tidak terpenuhi atau b <= 0.")
```

PRAKTIKUM 3: PERCABANGAN

```python
# 1. IF Sederhana
nilai = 85
print("Contoh IF Sederhana:")
if nilai > 80:
    print("Selamat! Anda lulus dengan nilai tinggi.\n")

# 2. IF-ELSE
umur = 17
print("Contoh IF-ELSE:")
if umur >= 18:
    print("Anda sudah cukup umur untuk mendapatkan SIM.\n")
else:
    print("Anda belum cukup umur untuk mendapatkan SIM.\n")

# 3. IF-ELIF-ELSE
hari = "Rabu"
print("Contoh IF-ELIF-ELSE:")
if hari == "Senin":
    print("Hari Senin: Saatnya kembali bekerja!")
elif hari == "Selasa":
    print("Hari Selasa: Jadwal rapat mingguan.")
elif hari == "Rabu":
    print("Hari Rabu: Ada diskon di beberapa toko.")
else:
    print("Hari lainnya: Atur jadwalmu dengan baik.\n")

# 4. IF Bersarang (Nested IF)
suhu = 35
print("\nContoh IF Bersarang (Nested IF):")
if suhu > 30:
    print("Cuaca cukup panas.")
    if suhu > 40:
        print("Bahkan sangat terik! Disarankan banyak minum air.")
    else:
        print("Masih relatif normal, tapi tetap jaga kesehatan.")
else:
    print("Cuaca sepertinya cukup sejuk.\n")

# 5. Menggabungkan Percabangan dengan Operasi Logika
nilai_teori = 75
nilai_praktik = 80
print("\nContoh IF dengan Operasi Logika AND/OR:")
if nilai_teori >= 70 and nilai_praktik >= 70:
    print("Anda lulus karena nilai teori dan praktik memadai.")
elif nilai_teori < 70 and nilai_praktik < 70:
    print("Anda perlu meningkatkan nilai teori dan praktik.")
elif nilai_teori < 70:
    print("Anda perlu meningkatkan nilai teori.")
else:
    print("Anda perlu meningkatkan nilai praktik.\n")

# 6. Penggunaan If Ternary
angka = -5
print("Contoh If Ternary (Conditional Expression):")
status = "Positif" if angka > 0 else "Negatif atau Nol"
print("Angka", angka, "=>", status)
```

PRAKTIKUM 4: PERULANGAN

```python
# 1. FOR Loop dengan range()
print("1) FOR loop dengan range()")
for i in range(5):
    print("Perulangan ke-", i)

# 2. FOR Loop mengiterasi List
print("\n2) FOR loop mengiterasi list")
buah = ["apel", "mangga", "jeruk", "pisang"]
for item in buah:
    print("Buah:", item)

# 3. WHILE Loop
print("\n3) WHILE loop sederhana")
count = 0
while count < 5:
    print("count =", count)
    count += 1

# 4. BREAK di dalam loop
print("\n4) BREAK di dalam loop")
for i in range(10):
    if i == 3:
        print("Loop dihentikan pada i =", i)
        break
    print("i =", i)

# 5. CONTINUE di dalam loop
print("\n5) CONTINUE di dalam loop")
for i in range(5):
    if i == 2:
        print("Lewati i =", i, "dengan continue")
        continue
    print("i =", i)

# 6. NESTED loop (Loop Bersarang)
print("\n6) NESTED loop")
for i in range(3):
    for j in range(2):
        print(f"i={i}, j={j}")

# 7. ELSE pada loop for/while
print("\n7) ELSE pada loop for/while")
for x in range(3):
    print("x =", x)
else:
    print("Loop for telah selesai tanpa break.")

y = 0
while y < 3:
    print("y =", y)
    y += 1
else:
    print("Loop while telah selesai tanpa break.")

# 8. PASS (placeholder)
print("\n8) PASS (placeholder)")
for i in range(3):
    if i == 1:
        pass # tidak melakukan apa-apa
    print("i =", i)
```

PENUGASAN 1:

```python
# Program 1: Menghitung BMI (Body Mass Index)
print("== Program Penghitung BMI ===")

# Input dari pengguna [cite: 479]
berat = float(input("Masukan berat badan (kg): "))
tinggi_cm = float(input("Masukan tinggi badan (cm): "))

# Konversi tinggi dari cm ke meter
tinggi_m = tinggi_cm / 100

# Hitung BMI menggunakan rumus
bmi = berat / (tinggi_m ** 2)

# Menentukan kategori berdasarkan hasil [cite: 478]
if bmi < 18.5:
    kategori = "Underwight (Kekurangan berat badan)"
elif 18.5 <= bmi <= 24.9:
    kategori = "Normal (Ideal)"
elif 25 <= bmi <= 29.9:
    kategori = "Overweight (Kelebihan berat badan)"
else:
    kategori = "Obesity (Kegemukan)"

# Output hasil [cite: 478]
print(f"\nNilai BMI Anda: {bmi:.2f}")
print(f"Kategori: {kategori}")
```

```python
# Program 2: Mengecek Bilangan Genap, Ganjil, dan Prima
print("=== Program Cek Sifat Bilangan ===")

angka = int(input("Masukkan sebuah bilangan bulat positif: "))

# 1. Cek Genap atau Ganjil menggunakan modulus (5) [cite: 73]
if angka % 2 == 0:
    sifat_dasar = "Genap"
else:
    sifat_dasar = "Ganjil"

# 2. Cek apakah bilangan Prima
# Bilangan prima adalah bilangan > 1 yanag habis dibagi 1 dan dirinya sendiri
is_prima = True
if angka < 2:
    is_prima = False
else:
    for i in range(2, int(angka**0.5) + 1): # Cek pembagi dari 2 hingga akar angka
        if angka % i == 0:
            is_prima = False
            break

# Output hasil
print(f"\nBilangan {angka} adalah bilangan {sifat_dasar}.")

if is_prima:
    print(f"{angka} juga merupakan bilangan prima.")
else:
    print(f"{angka} BUKAN merupakan bilangan prima.")
```
