Update & operasi dasar list

```python
nilai = [85, 90, 78]
nilai[1] = 95
nilai.append(100)
nilai.remove(78)
print(nilai)
```

Slicing list

```python
nilai = [85, 90, 78, 92, 88]
print(nilai[0:3])   # [85, 90, 78]
print(nilai[2:])    # [78, 92, 88]
print(nilai[3:])    # [85, 90, 78]
```

List 2D ( Matriks )

```python
matriks = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

print(matriks[0][0])  # 1
print(matriks[1][2])  # 6
print(matriks[2])     #  [7, 8, 9]
```

Operasi pada list 2D

```python
A = [[1, 2, 3], [4, 5, 6]]
B = [[7, 8, 9], [10, 11, 12]]
C = []

for i in range(len(A)):
    baris = []
    for j in range(len(A[0])):
      baris.append(A[i][j] + B[i][j])
    C.append(baris)
print (C)
```

Studi kasus : Pengolahan nilai mahasiswa

```python
nilai = [
    [80, 90, 85],
    [70, 75, 80],
    [88, 92, 84]
]

for i, baris in enumerate(nilai, start=1):
    rata = sum(baris) / len(baris)
    print(f"Rata-rata mahasiswa {i}: {rata:.2f}")
```

Operasi matriks 2D

```python
matriks = [
    [1, 2],
    [3, 4]
]
matriks2 = [
    [5, 6],
    [7, 8]
]
hasil = [
    [matriks[0][0] + matriks2[0][0], matriks[0][1] + matriks2[0][1]],
    [matriks[1][0] + matriks2[1][0], matriks[1][1] + matriks2[1][1]]
]
print("Hasil penjumlahan matriks:")
for baris in hasil:
  print(baris)
```

Rata-rata nilai mahasiswa

```python
nilai = [
    [75, 80, 85],
    [60, 70, 65,],
    [90, 95, 92]
]
for i in range(len(nilai)):
  rata = sum(nilai[i]) / len(nilai[i])
  print(f"Rata-rata mahasiswa {i+1}: {rata:.2f}")
```

Studi kasus : Data toko

```python

Stok = [
    ["Buku", 20],
    ["Pulpen", 35],
    ["Penghapus", 15]
]
print ("Daftar Stok Barang")
for item in Stok:
  print(f"- {item[0]}: {item[1]} buah")

barang_terbanyak = max(Stok, key=lambda x: x[1])
print("\nBarang dengan stok terbanyak:")
print(f"{barang_terbanyak[0]} ({barang_terbanyak[1]} buah)")
```
