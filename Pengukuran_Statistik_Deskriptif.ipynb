1. Central Tendency

```python
# Mendapatkan nilai Mean
import numpy as np

# Data nilai
score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

# Menghitung rata-rata (mean)
x = np.mean(score)
print(f"Mean: {x}")
```

```python
# Mendapatkan nilai Median
import numpy as np

# Data nilai
score = [90, 80, 70, 65, 110, 80, 83, 67, 94]

# Menghitung nilai tengah (median)
x = np.median(score)
print(f"Median: {x}")
```

```python
# Mendapatkan nilai Modus / Mode
from scipy import stats

# Data nilai
score = [90,80,70,5,110,80,83,67,94]

# Menghitung modus
x = stats.mode(score, keepdims=True)
print(f"Modus: {x.mode[0]}")
print(f"Jumlah kemuculan: {x.count[0]}")
```

2. Spread (Sebaran Data)

```python
# Mendapatkan Nilai Range
import numpy as np

# Data
data = [2, 3, 7, 8, 9, 12, 15, 20, 50, 42]

# Menghitung Range (Jangkauan)
minimum = np.min(data)
maximum = np.max(data)
x = maximum - minimum

print(f"Nilai Min: {minimum}")
print(f"Nilai Max: {maximum}")
print(f"Range: {x}")
```

b.	Mendapatkan Nilai Percentile

```python
import numpy as np

# Data
data = [90,80,70,65,110,80,83,67,94]

# Menghitung Persentil ke-90
x = np.percentile(data, 90)
print(f"Persentil ke-90: {x}")
```

Mendapatkan Nilai Quartile dan Interquartile Range (IQR)

```python
import numpy as np

# Data
data = [3,4,4,5,5,6,6,7,8,8,9,10]

# Menghitung Q1,Q3 dan IQR
Q1 = np.median(data[:int(len(data)/2)])
Q3 = np.median(data[int(len(data)/2):])
IQR = Q3 - Q1

print(f"Q1: {Q1}")
print(f"Q3: {Q3}")
print(f"IQR: {IQR}")
```

Mendapatkan Nilai Variance dan Standard Deviation

```python
import numpy as np

# Data
data = [86, 87, 88, 86, 87, 85, 86]

# Menghitung Mean, Varians, dan Standar Deviasi
mean = np.mean(data)
var = np.var(data)
std = np.std(data)

print(f"Mean: {mean}")
print(f"Variance: {var}")
print(f"Standard Deviation: {std}")
```

```python
import numpy as np
from scipy import stats

# Data pelemparan dadu 25 kali [cite: 89, 90]
data_dadu = [1, 2, 3, 4, 5, 5, 6, 2, 3, 4, 5, 6, 6, 4, 3, 2, 1, 4, 3, 5, 6, 6, 5, 4, 5]
modus_dadu = stats.mode(data_dadu, keepdims=True)

print("=== NOMOR 1 ===")
print(f"Modus data dadu: {modus_dadu.mode[0]} (muncul sebanyak {modus_dadu.count[0]} kali)")
print("-" * 30)
```

```python
# Deret data [cite: 92]
data_deret = [9, 10, 11, 6, 8, 7, 7, 5, 4, 5]
p25 = np.percentile(data_deret, 25)
p75 = np.percentile(data_deret, 75)

print("=== NOMOR 2 ===")
print(f"Persentil ke-25: {p25}")
print(f"Persentil ke-75: {p75}")
print("-" * 30)
```

```python
# Data hasil ujian [cite: 93]
nilai_ujian = [50, 54, 62, 50, 52, 59, 61, 63, 65, 10, 53, 63, 65, 50, 59, 62, 50, 51, 57, 60, 63, 65, 65, 53, 99]

# 94. Urutkan data
nilai_urut = sorted(nilai_ujian)

# 95. Identifikasi Q1, Q3, dan IQR [cite: 62, 95]
# Menggunakan metode numpy untuk akurasi data statistik
Q1 = np.percentile(nilai_urut, 25)
Q3 = np.percentile(nilai_urut, 75)
IQR = Q3 - Q1

# 96. Upper bound dan Lower bound [cite: 96]
lower_bound = Q1 - (1.5 * IQR)
upper_bound = Q3 + (1.5 * IQR)

# 97. Identifikasi Outlier [cite: 97]
outliers = [x for x in nilai_urut if x < lower_bound or x > upper_bound]

print("=== NOMOR 3 ===")
print(f"Data Berurutan: {nilai_urut}")
print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
print(f"Lower Bound: {lower_bound}")
print(f"Upper Bound: {upper_bound}")
print(f"Data Outlier: {outliers}")
print("-" * 30)
```

```python
# Data dari tabel [cite: 99]
# Berat Badan (Range) -> Titik Tengah (xi)
# 41-45 -> 43
# 46-50 -> 48
# 51-55 -> 53 (Koreksi dari 50-55 di teks agar konsisten 5 poin)
# 56-60 -> 58
# 61-65 -> 63
# 66-70 -> 68

xi = np.array([43, 48, 53, 58, 63, 68])
fi = np.array([2, 3, 6, 4, 5, 2])

# Hitung Mean Data Kelompok
mean_kelompok = sum(fi * xi) / sum(fi)

# Hitung Varians (s^2) dan Standar Deviasi (s)
varians_kelompok = sum(fi * (xi - mean_kelompok)**2) / (sum(fi) - 1)
std_dev_kelompok = np.sqrt(varians_kelompok)

print("=== NOMOR 4 ===")
print(f"Rata-rata Berat Badan: {mean_kelompok:.2f} kg")
print(f"Standar Deviasi: {std_dev_kelompok:.2f}")
```
